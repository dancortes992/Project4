```python
# Import libraries
import pandas as pd
import plotly_express as px
```

```python
# Read data
# Copy data 
vehicle_data = pd.read_csv(r"C:\Users\danco\Documents\Tripleten\MyProjects\Project4\vehicles_us.csv")
df = vehicle_data.copy()
df.info()
```

```python
# View dataframe
display(df)
```

```python
# Correct data types
df['date_posted'] = pd.to_datetime(df['date_posted'], format= '%Y-%m-%d')
df['cylinders'].fillna(0, inplace=True)
df['odometer'].fillna(0, inplace=True)
df['paint_color'].fillna('N/A', inplace=True)
df['is_4wd'].fillna(0, inplace=True)
df['model_year'].fillna(0, inplace=True)
df['model_year'] = df['model_year'].astype(int)
df['model_year'] = df['model_year'].astype(str)
df['model_year'] = df['model_year'].replace('0', 'N/A')


# Add features: Manufacturer column
df['manufacturer'] = df['model'].str.split(' ').str[0]
df['model'] = df['model'].str.split(' ').str[1]
```

```python
# Aggregate data
df_make_type = df.groupby(['manufacturer', 'type'])['model'].count().reset_index()
df_make_type.columns = ['manufacturer', 'type', 'count']
```

```python
# Vehicle types by manufaturer bar chart
df_make_type_bar = px.bar(df_make_type, x="manufacturer", y="count", color="type", title="Vehicle Types by Manufacturer")
df_make_type_bar.show()
```

```python
#Vehicle condition vs model year histogram
df_cond_year_hist = px.histogram(df, x="model_year", color="condition", title="Vehicle Condition vs. Model Year",
                                 labels=dict(model_year="Model year"))
df_cond_year_hist.show()
```

```python
#Vehicle Year vs. price scatterplot
df_price_year_scat = px.scatter(df, x='model_year', y='price', color='model', title="Vehicle year vs. Price",
                                labels=dict(price="Price", model_year="Model year"))
df_price_year_scat.show()
```
